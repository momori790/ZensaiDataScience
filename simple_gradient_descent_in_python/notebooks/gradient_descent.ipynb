```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10,8))
x = np.arange(np.pi/2, 5*np.pi/2, 0.01)
y = np.sin(x)

plt.plot(x, y)
```

```python
def plot_slope_at_point(x):
    ''' plot sin graph and steepest slope at point x '''
    plt.figure(figsize=(10,8))
    
    # plots line of steepest slope at point x
    y = np.sin(x)
    slope = np.cos(x)
    b = y - slope * x
    
    x_range = np.arange(x - np.pi/3, x + np.pi/3, 0.01)
    
    plt.plot(x_range, x_range * slope + b)
    
    # plot tangent point
    plt.plot(x, y, 'ro')
    
    # plot the sin graph
    x_range = np.arange(np.pi/2, 5 * np.pi/2, 0.01)
    y = np.sin(x_range)

    plt.plot(x_range, y)
```

```python
plot_slope_at_point(np.pi)
```

```python
def gradient_descent(starting_point, step_size=0.7, num_iter=10):
    ''' plot the progress of gradient descent algorithm '''
    fig, axes = plt.subplots(2, 5, figsize=(18,8))
    
    for i in range(num_iter):
        # build the axes for displaying the progress
        row = int(i/5)
        col = i%5
        ax = axes[row][col]

        # plots line of steepest slope at point x
        y = np.sin(starting_point)
        slope = np.cos(starting_point)
        b = y - slope * starting_point

        x_range = np.arange(starting_point - np.pi/3, starting_point + np.pi/3)
        ax.plot(x_range, x_range * slope + b)
        
        # plot tangent point
        ax.plot(starting_point, y, 'ro')

        # plot the sin graph
        x_range = np.arange(np.pi/2, 5 * np.pi/2, 0.01)
        y = np.sin(x_range)

        ax.plot(x_range, y)
        ax.set_title('iteration {}'.format(i+1))
        
        # update new x value
        starting_point = starting_point - step_size*slope
```

```python
# Good example
gradient_descent(2*np.pi/3)
```

```python
# Step size is too big
gradient_descent(2*np.pi/3, step_size=2)
```

```python
# Step size is too small
gradient_descent(2*np.pi/3, step_size=0.1)
```
