```python
# There seems to be some correlation between living area space to price
fig, ax = plt.subplots(figsize=(10,7))
ax.scatter(df.GrLivArea, df.SalePrice)

ax.set_ylabel('price')
ax.set_xlabel('Living Area Space')
ax.set_title('Affect of living area on price')
plt.show()
```

```python
class LinearRegression(object):
    '''linear regression object.'''
    def __init__(self, num_iter=15, lr=0.00000005):
        self.num_iter = num_iter
        self.lr = lr
                
    def fit(self, x, y):
        '''
        x is a nxm matrix, and y a nx1 array.
        '''        
        
        # Set up the visualization
        plt.figure(figsize=(20,10))
        
        #init weights
        self.theta = np.zeros(x.shape[1])
        self.b = 0

        for i in range(self.num_iter):
            predictions = self.predict(x)
            #calculate gradient
            gradient_theta = -2*np.dot(x.T, (y-predictions))/y.size
            gradient_b = -2*(sum(y-predictions))/y.size
            
            # Visualize the regression line every 2 iterations
            if i%2==0:
                plt.plot(x, self.theta*x + self.b, label='iteration: {}'.format(i))
                
            # update the weights
            self.theta -= self.lr * gradient_theta
            self.b -= self.lr * gradient_b
            
        # overlay scatter plot on figure
        plt.scatter(df.GrLivArea, df.SalePrice)
        plt.ylabel('price')
        plt.xlabel('Living Area Space')
        plt.title('Effect of living area on price')    
        plt.legend()
            
      
    def predict(self, x):
        return np.dot(x, self.theta) + self.b
    

    def describe(self):
        print('slope: {}'.format(self.theta))
        print('intercept: {}'.format(self.b))
        
        
```

```python
lr = LinearRegression()
lr.fit(df['GrLivArea'].values.reshape(1460,1), df['SalePrice'])
```
